### loc・ilocの使い方

```python
df=pd.DataFrame(np.arange(20).reshape((5,4)),
                index=["1行目","2行目","3行目","4行目","5行目"],
                columns=["A列","B列","C列","D列"])
df
```

```python
df.loc[:,"A列"]
```

```python
df.loc[:,["A列","C列"]]
```

```python
df.loc["3行目",:]
```

```python
df.loc[["2行目","4行目",],:]
```

```python
df.loc[["2行目","5行目"],["B列","C列"]]
```

```python
df.iloc[0,2]
```

```python
df.iloc[2:,3]
```

```python
df.iloc[1:,:2]
```
