```python
import matplotlib.pyplot as plt

x=[1,2,3]
y=[2,4,6]

plt.plot(x,y)
plt.show()
```

```python
x = [1, 2, 3]
y = [2, 4, 6]

fig, ax = plt.subplots()
plt.show()
```

```python
x1 = [1, 2, 3]
y1 = [2, 4, 6]
x2 = [1, 2, 3]
y2 = [-2, -4, -6]

fig,axes=plt.subplots(2)
axes[0].plot(x1,y1)
axes[1].plot(x2,y2)
plt.show
```

```python
fig, ax = plt.subplots(2, 2)
plt.show()
```

```python
fig, ax = plt.subplots(ncols=2)
plt.show()
```

```python
matplotlib.style.use('seaborn')

x = [1, 2, 3]
y = [2, 4, 6]

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

```python
matplotlib.style.use('dark_background')

x = [1, 2, 3]
y = [2, 4, 6]

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

```python
x1 = [1, 2, 3]
y1 = [2, 4, 6]
x2 = [1, 2, 3]
y2 = [-2, -4, -6]

fig, axes = plt.subplots(2)
axes[0].plot(x1, y1)
axes[1].plot(x2, y2)
axes[0].set_title('Subplot1 title')
axes[1].set_title('Subplot2 title')
fig.suptitle('Graph title')
plt.show()
```

```python
x = [1, 2, 3]
y = [2, 4, 6]

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel('X lable')
ax.set_ylabel('Y lable')
plt.show()
```

```python
x = [1, 2, 3]
y = [2, 4, 6]

fig, ax = plt.subplots()
ax.plot(x, y, label='sample legend')
ax.legend()
plt.show()
```

```python
x = [1, 2, 3]
y = [2, 4, 6]

fig, ax = plt.subplots()
ax.plot(x, y, label='sample legend')
ax.legend(loc='lower right')
plt.show()
```
